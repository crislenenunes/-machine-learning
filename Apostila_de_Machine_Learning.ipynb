# Modelos de Aprendizado de Máquina - (Algoritmos) - **Superviosionados**

### Principais tipos de algoritmos de aprendizado de máquina

**Regressão Linear**  
Usado para prever valores numéricos.  
*Exemplo:* Estimar o preço de uma casa com base no tamanho, localização, número de quartos, etc.

---

**Regressão Logística**  
Usado para classificar em dois grupos (sim/não, verdadeiro/falso).  
*Exemplo:* Prever se um e-mail é spam ou não.

---

**Árvores de Decisão**  
Funciona como um fluxograma de perguntas "sim ou não" para chegar a uma conclusão.  
*Exemplo:* Prever se um cliente vai cancelar um serviço com base no comportamento dele.

---

**Random Forest**  
Combina várias árvores de decisão para melhorar a precisão das previsões.  
*Exemplo:* Usado em diagnósticos médicos ou para detectar fraudes financeiras.

---

**K-Nearest Neighbors (KNN)**  
Classifica algo com base nos "vizinhos" mais próximos — ou seja, exemplos semelhantes.  
*Exemplo:* Classificar um animal com base em características parecidas de outros animais conhecidos.

---

**Support Vector Machine (SVM)**  
Cria uma linha (ou plano) que separa os grupos da melhor forma possível.  
*Exemplo:* Identificar rostos em fotos, separando rostos conhecidos dos desconhecidos.

### Principais tipos de algoritmos de agrupamento e redução de dimensionalidade

**K-Means**  
Agrupa os dados em *K* grupos com características semelhantes.  
Exemplo: Segmentar clientes com base no perfil de consumo.

---

**Hierarchical Clustering (Agrupamento Hierárquico)**  
Cria uma “árvore” de agrupamentos, partindo de um grupo geral até subdivisões mais específicas.  
Exemplo: Classificar músicas por estilo, gênero e subgêneros.

---

**DBSCAN (Density-Based Spatial Clustering)**  
Agrupa pontos que estão próximos uns dos outros e ignora os pontos isolados (outliers).  
Exemplo: Detectar comportamentos fora do padrão, como fraudes ou acessos incomuns.

---

**PCA (Principal Component Analysis)**  
Reduz a quantidade de variáveis dos dados, mantendo apenas as informações mais relevantes.  
Exemplo: Simplificar conjuntos de dados com muitas colunas, como em imagens ou sensores.

### O que é Machine Learning?

Antes de entender como funciona a regressão linear, é importante compreender o conceito de *machine learning* (aprendizado de máquina).

Em vez de programar o computador com uma fórmula fixa para resolver um problema, você fornece a ele diversos exemplos reais — chamados de dados. A partir desses exemplos, o computador é capaz de aprender padrões por conta própria e criar uma forma de realizar previsões.

Por exemplo:  
Se quisermos que o computador preveja o preço de uma casa com base no tamanho, não precisamos dizer a fórmula exata. Basta mostrar diversos exemplos de casas com seus respectivos tamanhos e preços, e ele aprende a fazer essa previsão sozinho.

---

### O que é Regressão Linear?

A regressão linear é um dos algoritmos mais simples — e também um dos mais fundamentais — do aprendizado de máquina.  
Ela é usada para prever um valor numérico (também chamado de valor contínuo) com base em uma ou mais variáveis de entrada.

**Exemplo prático:**  
Você tem várias informações de casas como estas:

| Tamanho (m²) | Preço (mil R$) |
|--------------|----------------|
| 50           | 200            |
| 80           | 320            |
| 120          | 480            |

Com base nesses dados, a regressão linear tenta encontrar uma relação entre o tamanho e o preço — uma linha reta que represente essa tendência. Assim, ao receber o tamanho de uma nova casa, o modelo consegue prever o preço aproximado com base no padrão aprendido.

---

### Como funciona a Regressão Linear?

Vamos supor que você tenha essa dúvida:

> “Se eu tiver uma casa com 75 m², qual será o preço dela?”

A regressão linear tenta resolver exatamente esse tipo de problema, encontrando uma **linha reta** que melhor se ajusta aos dados para fazer previsões.

---

### A fórmula da linha

A linha da regressão segue uma fórmula muito parecida com aquela da matemática básica:

\[
y = a \cdot x + b
\]

No nosso exemplo:

\[
\text{preço} = a \cdot \text{tamanho} + b
\]

- **x**: valor de entrada (por exemplo, o tamanho da casa)  
- **y**: valor que queremos prever (neste caso, o preço)  
- **a**: inclinação da linha – mostra quanto o preço aumenta a cada metro quadrado a mais  
- **b**: interceptação – é o ponto onde a linha cruza o eixo y (ou seja, o preço base quando o tamanho é zero)

---

### Como o computador aprende?

O algoritmo de regressão linear **aprende** a encontrar os melhores valores de `a` e `b` com base em dados reais. O processo é assim:

1. Você fornece exemplos reais (como tamanho e preço de várias casas).
2. O algoritmo começa com um chute inicial para `a` e `b`.
3. Compara o valor previsto com o valor real e calcula o erro.
4. Ajusta um pouco os valores de `a` e `b` para reduzir o erro.
5. Repete esse processo várias vezes até encontrar a melhor linha possível.

Esse processo é chamado de **treinamento**.

---

### Visualmente, como isso aparece?

Se colocarmos os dados em um gráfico:

- O eixo **X** representa o tamanho da casa (entrada). **meus dados**
- O eixo **Y** representa o preço (saída). **Quem eu estou tentando prever**
- Cada casa vira um **ponto** no gráfico.
- A regressão linear traça uma **linha reta** que tenta passar o mais próximo possível de todos esses pontos — minimizando os erros.

---

###Exemplo 1:

````python
# Passo 1: Dados de entrada
# Temos os tamanhos das casas (em metros quadrados) e seus respectivos preços (em reais)
tamanho = [50, 60, 70, 80]
preco = 100000, 120000, 140000, 160000

# Passo 2: Preparando os dados para o modelo

# O modelo de regressão linear do Scikit-learn espera os dados em um formato específico: uma matriz (2 dimensões).
# Por isso, usamos o NumPy para transformar a lista de tamanhos em uma matriz de uma coluna.

tamanho = np.array(tamanho).reshape(-1, 1)

'''
np.array(tamanho): converte a lista simples em um array NumPy, que permite operações matemáticas mais eficientes.

.reshape(-1, 1): transforma esse array em uma matriz com uma coluna e "N" linhas (quantidade de dados).
O "-1" é um atalho que diz: "descubra automaticamente quantas linhas são necessárias".
Assim, o modelo entende que temos uma única variável de entrada: o tamanho da casa.
'''

# Passo 3: Criando e treinando o modelo

# Agora vamos criar um modelo de regressão linear
modelo = LinearRegression()

# Com .fit(), treinamos o modelo com os dados fornecidos:
# Ele vai encontrar os melhores valores para os coeficientes (a e b) da equação da linha: y = a*x + b
modelo.fit(tamanho, preco)

'''
LinearRegression(): cria um objeto do modelo de regressão linear.

.fit(tamanho, preco): ensina o modelo a "enxergar" a relação entre o tamanho da casa (entrada) e o preço (saída).
Nesse processo, o modelo encontra a linha que melhor se ajusta aos dados (minimizando os erros).
'''

# Passo 4: Gerando os valores para desenhar a linha de regressão no gráfico

# Para que o gráfico fique suave, vamos gerar 100 tamanhos entre o menor e o maior tamanho da casa.
tamanho_para_plotar = np.linspace(tamanho.min(), tamanho.max(), 100).reshape(-1, 1)

'''
np.linspace(tamanho.min(), tamanho.max(), 100): cria 100 valores igualmente espaçados entre 50 e 80.
Esses valores vão nos ajudar a desenhar uma linha reta contínua no gráfico.

.reshape(-1, 1): novamente, transformamos em uma matriz com 1 coluna, como o modelo espera.
'''

# Agora usamos o modelo para prever os preços para esses 100 tamanhos.
preco_previsto = modelo.predict(tamanho_para_plotar)

'''
modelo.predict(...): o modelo aplica a fórmula aprendida (y = a*x + b) a cada um dos 100 tamanhos e retorna os preços previstos.
Esses valores formarão a linha vermelha no gráfico.
'''

# Passo 5: Visualizando os dados e o resultado da regressão linear

plt.figure(figsize=(10, 6))  # Define o tamanho do gráfico

# Mostra os dados reais (tamanhos e preços das casas) como pontos azuis
plt.scatter(tamanho, preco, color='blue', label='Casas (dados reais)')

# Desenha a linha vermelha da regressão com os preços previstos
plt.plot(tamanho_para_plotar, preco_previsto, color='red', linewidth=2, label='Linha de regressão')

# Personalização dos eixos e título
plt.xlabel('Tamanho da casa (m²)')         # Legenda do eixo X
plt.ylabel('Preço da casa (R$)')           # Legenda do eixo Y
plt.title('Regressão Linear: Preço vs. Tamanho da Casa')  # Título do gráfico
plt.legend()     # Mostra a legenda com rótulos das curvas
plt.grid(True)   # Adiciona uma grade de fundo para facilitar a leitura
plt.show()       # Exibe o gráfico
````

````python
# Avaliando o desempenho do modelo com o coeficiente de determinação (R²)
r2 = modelo.score(tamanho, preco)

'''
modelo.score(tamanho, preco):
Calcula o R² (coeficiente de determinação), que mede o quanto a linha de regressão explica a variação dos dados.

Valores possíveis:
- R² = 1.0: ajuste perfeito (a linha passa exatamente por todos os pontos)
- R² = 0.0: o modelo não explica nada da variação nos dados
- Quanto mais próximo de 1, melhor é o desempenho do modelo

Nesse caso, como usamos os mesmos dados de treino, o valor será alto (ajuste quase perfeito),
mas o ideal é calcular o score com dados de teste (que o modelo nunca viu) para avaliar melhor sua capacidade de generalização.
'''
print(f"R² do modelo: {r2:.2f}")
````

###Exemplo 2:

````python
'''
Considerando uma base com dados sobre investimento em propaganda e vendas,
vamos repetir o processo de regressão linear.
'''

# Dados fictícios: valores investidos em propaganda e respectivas vendas
propaganda = [1, 2, 3, 4, 5]         # Valores de entrada (R$ investido em propaganda)
vendas = [1, 2, 4, 5, 8]             # Valores de saída (R$ em vendas)

# Transformando os dados para o formato que o modelo aceita
propaganda = np.array(propaganda).reshape(-1, 1)

'''
np.array(propaganda): transforma a lista em um array do NumPy, que é aceito pelo modelo.
.reshape(-1, 1): reorganiza o array para virar uma matriz com 1 coluna (formato necessário para o modelo).
- O "-1" diz ao NumPy para calcular automaticamente o número de linhas necessário.
'''

# Criando e treinando o modelo
modelo = LinearRegression()
modelo.fit(propaganda, vendas)

'''
LinearRegression(): cria um modelo de regressão linear.
modelo.fit(propaganda, vendas): o modelo "aprende" a relação entre os valores de propaganda e as vendas.
Ou seja, ele encontra os coeficientes da reta que melhor se ajusta aos dados.
'''

# Preparando os dados para plotar a linha de regressão
tamanho_para_plotar2 = np.linspace(propaganda.min(), propaganda.max(), 100).reshape(-1, 1)
vendas_previstas = modelo.predict(tamanho_para_plotar2)

'''
np.linspace(...): cria 100 valores entre o menor e o maior valor de propaganda.
Isso ajuda a gerar uma linha suave para o gráfico.

modelo.predict(...): usa o modelo treinado para prever as vendas com base nos valores de propaganda.
O resultado será uma sequência de pontos que formarão a linha de regressão.
'''

# Visualizando os dados e a linha de regressão
plt.figure(figsize=(10, 6))
plt.scatter(propaganda, vendas, color='blue', label='Vendas (dados reais)')
plt.plot(tamanho_para_plotar2, vendas_previstas, color='red', linewidth=2, label='Linha de regressão')
plt.xlabel('Investimento em Propaganda (R$)')
plt.ylabel('Vendas (R$)')
plt.title('Regressão Linear: Vendas vs. Propaganda')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Coeficientes da equação da reta de regressão
'''
O modelo encontra uma equação no formato:

y = a * x + b

No nosso exemplo:
vendas = a * propaganda + b

Onde:
- a é o coeficiente angular (mostra o quanto as vendas aumentam a cada unidade de propaganda)
- b é o intercepto (valor de vendas quando o investimento em propaganda é zero)

Esses valores são aprendidos automaticamente durante o treinamento com .fit()
'''

print("Coeficiente a (inclinação da reta):", modelo.coef_)        # Mostra o valor de 'a'
print("Intercepto b (onde a reta cruza o eixo Y):", modelo.intercept_)  # Mostra o valor de 'b'
print("R² do modelo:", modelo.score(propaganda, vendas))          # Avalia a qualidade do modelo (0 a 1)

'''
- modelo.coef_: retorna o valor do coeficiente angular "a".
- modelo.intercept_: retorna o valor do intercepto "b".
- modelo.score(...): retorna o R² (coeficiente de determinação), que mede o quanto a reta explica a variação dos dados.

Quanto mais próximo de 1 o R², melhor é o ajuste do modelo aos dados.
'''
````

## Statsmodels (SM)

````python
import statsmodels.api as sm

# Preparando os dados
x = np.array(propaganda)         # Transforma a lista de propaganda em array NumPy
x = sm.add_constant(x)           # Adiciona uma coluna de 1s ao array (necessária para calcular o intercepto b)
y = np.array(vendas)             # Transforma a lista de vendas em array NumPy

'''
sm.add_constant(x):
Adiciona uma constante ao conjunto de dados x.
Isso é necessário para que o modelo calcule o intercepto (b) da equação da reta.
Sem essa constante, o modelo assume que a linha passa pela origem (b = 0).
'''

# Criando e ajustando o modelo de regressão linear com statsmodels
modelo_stats = sm.OLS(y, x).fit()

'''
sm.OLS(y, x): cria um modelo de Regressão Linear (OLS = Ordinary Least Squares = Mínimos Quadrados Ordinários). Calcula a diferença de cada ponto da reta.
.fit(): treina o modelo com os dados fornecidos (como no scikit-learn).
'''

# Exibindo o resumo estatístico completo
print(modelo_stats.summary())

''' modelo_stats.summary():
Gera um relatório completo com diversas informações estatísticas sobre o modelo.

Principais itens do relatório:
- coef: coeficientes da reta (inclinação e intercepto)
- std err: erro padrão dos coeficientes
- t: valor do teste t para cada coeficiente
- P>|t|: valor-p (verifica se o coeficiente é estatisticamente significativo)
- R-squared: coeficiente de determinação R²
- Adj. R-squared: R² ajustado (corrige R² para múltiplas variáveis)
- F-statistic: testa se o modelo como um todo é significativo
- Prob (F-statistic): valor-p do F
- AIC e BIC: critérios de qualidade do modelo (menores = melhor)

Esse resumo é muito usado em análise estatística para avaliar se a regressão é confiável.'''
````

# Interpretação da Regressão Linear com `statsmodels`

### Medidas Gerais

- **Dep. Variable: y**  
  → É a variável dependente, ou seja, a que estamos tentando prever. No exemplo, são as **vendas**.

- **Model: OLS**  
  → O modelo utilizado foi **OLS (Ordinary Least Squares)**, ou seja, **Regressão Linear por Mínimos Quadrados Ordinários**.

- **Method: Least Squares**  
  → É o método matemático usado para ajustar a reta da regressão, tentando **minimizar a soma dos quadrados dos erros** entre os valores reais e os previstos.

- **No. Observations: 5**  
  → Temos **5 observações** (5 pares de valores de propaganda e vendas usados para treinar o modelo).

- **Df Residuals: 3**  
  → Os **graus de liberdade dos resíduos** são 3. Isso é calculado como:  
  `número de observações - número de parâmetros estimados`  
  Ou seja: 5 - 2 = 3 (porque o modelo estima 2 parâmetros: o intercepto e o coeficiente da variável `propaganda`).

- **Df Model: 1**  
  → Há **1 variável explicativa** no modelo, neste caso, a variável `propaganda`.

---

### Qualidade do Modelo

- **R-squared: 0.963**  
  → O **R² (coeficiente de determinação)** mede quanto da variação da variável dependente (vendas) é explicada pelo modelo.  
  Neste caso, o modelo explica **96,3% da variação das vendas** com base nos dados de propaganda. Isso indica um ajuste muito bom.

- **Adj. R-squared: 0.951**  
  → O **R² ajustado** é uma versão do R² que leva em conta o número de variáveis do modelo.  
  Como só temos uma variável, a diferença é pequena, mas é útil em modelos maiores.

- **F-statistic: 78.82**  
  → Mede a qualidade geral do modelo. Quanto maior o valor da estatística F, melhor o ajuste.

- **Prob (F-statistic): 0.003**  
  → Este é o **valor-p do teste F**. Indica a **probabilidade do modelo ser fruto do acaso**.  
  Como é **menor que 0.05**, concluímos que o modelo é **estatisticamente significativo**.  
  Isso significa que a variável `propaganda` realmente explica a variação nas vendas com **confiança estatística**.

---

### Coeficientes da Equação

A equação da reta estimada pelo modelo é:

```
vendas = -1.10 + 1.70 * propaganda
```

Ou, de forma geral:

```
y = a + b * x
```

- **const: -1.1000**  
  → É o **intercepto da reta**, ou seja, o valor previsto para as vendas quando a propaganda for zero.  
  O modelo prevê vendas = -1,10 nesse caso, o que **não faz sentido prático** (vendas negativas), mas é comum em modelos simples.

- **x1: 1.7000**  
  → É o **coeficiente da variável `propaganda`**. Indica que **para cada unidade a mais investida em propaganda, as vendas aumentam em média 1,70 unidades**.

#### Significância dos coeficientes:

- Para `const`, o **p-valor = 0.182** (maior que 0.05)  
  → Não é estatisticamente significativo. O intercepto pode ser zero sem afetar o modelo.

#**Regressão Múltipla**

### **Regressão Linear Múltipla: o que é e como funciona**

Quando queremos **estimar um valor (y)** com base em **duas ou mais variáveis independentes**, estamos usando o que chamamos de **Regressão Linear Múltipla**.

Vamos por partes:

#### **Regressão Linear Simples**
Na regressão linear simples, trabalhamos com **apenas uma variável independente**.  
Por exemplo: estimar o preço de uma casa (y) com base apenas no **tamanho da casa (x)**.

**Fórmula da regressão simples:**

\[
y = a \cdot x + b
\]

- \(a\): coeficiente (inclinação da reta)
- \(b\): intercepto (valor de y quando x = 0)

---

#### **Regressão Linear Múltipla**
Já na regressão linear múltipla, usamos **duas ou mais variáveis independentes** para explicar y.  
Por exemplo: estimar o preço de uma casa com base no **tamanho**, **número de quartos** e **localização**.

**Fórmula da regressão múltipla:**

\[
y = a_1 \cdot x_1 + a_2 \cdot x_2 + a_3 \cdot x_3 + \ldots + b
\]

Cada \(x_i\) representa uma variável independente, com seu respectivo coeficiente \(a_i\).

---

#### **Exemplo prático:**
Imagine que temos uma planilha com os seguintes dados de imóveis:

| Tamanho (m²) | Quartos | Bairro     | Preço (R$)   |
|--------------|---------|------------|--------------|
| 80           | 2       | Centro     | 400.000      |
| 100          | 3       | Subúrbio   | 350.000      |
| 120          | 3       | Centro     | 500.000      |
| 60           | 1       | Periferia  | 200.000      |

Nosso objetivo é criar um modelo que estime o preço (y) com base em múltiplas características da casa (X).

Esse tipo de regressão permite uma **previsão mais precisa**, já que considera vários fatores ao mesmo tempo, refletindo melhor a realidade.

image.png

### **Como funciona a Regressão Linear Múltipla**

Queremos prever o **preço de uma casa (y)** com base em múltiplas características:

- **Tamanho da casa** (em m²)  
- **Número de quartos**  
- **Distância do centro da cidade** (em km)

---

### **Fórmula Geral da Regressão Linear Múltipla**

\[
y = a_1 \cdot x_1 + a_2 \cdot x_2 + a_3 \cdot x_3 + \ldots + a_n \cdot x_n + b
\]

**Onde:**

- \(x_1, x_2, ..., x_n\) são as variáveis independentes (as entradas)
- \(a_1, a_2, ..., a_n\) são os coeficientes que o modelo aprende
- \(b\) é o **intercepto** (valor de y quando todos os x forem zero)

---

### **No nosso exemplo:**

\[
\text{Preço} = a_1 \cdot \text{Tamanho} + a_2 \cdot \text{Quartos} + a_3 \cdot \text{Distância} + b
\]

---

### **Como o computador aprende os valores de \(a_1, a_2, a_3\) e \(b\)?**

Esse processo é chamado de **treinamento** do modelo. O passo a passo é:

1. **Você fornece os dados**, com entradas (X) e saídas (y reais).  
2. O algoritmo começa com **valores iniciais aleatórios** para os coeficientes.  
3. Calcula os valores previstos para y com base nesses coeficientes.  
4. Compara os valores previstos com os reais, usando uma **função de erro** (ex: erro quadrático médio).  
5. Ajusta os coeficientes para **diminuir o erro**.  
6. Repete esse processo diversas vezes, até encontrar os **melhores coeficientes possíveis**.

---

### **Visualmente falando...**

- Com **1 variável independente**, o modelo encontra **uma linha** no plano (gráfico 2D).  
- Com **2 variáveis independentes**, o modelo forma **um plano** (gráfico 3D).  
- Com **mais de 2 variáveis**, temos um **hiperplano** (difícil de visualizar, mas matematicamente funciona da mesma forma).

---

### **Exemplo com valores hipotéticos**

Se o modelo chegar à seguinte equação:

\[
\text{Preço} = 1500 \cdot \text{Tamanho} + 20000 \cdot \text{Quartos} - 10000 \cdot \text{Distância} + 50000
\]

Podemos interpretar assim:

- Cada metro quadrado **aumenta o preço em R$1.500**  
- Cada quarto adicional **aumenta R$20.000**  
- Cada quilômetro distante do centro **reduz R$10.000**  
- Mesmo sem nenhuma dessas características, o preço **começa em R$50.000** (intercepto)



#Cuidados essenciais ao usar Regressão Linear

##1. Relação linear entre as variáveis##
A regressão linear só funciona bem quando há uma relação linear entre as
variáveis.

✅ O que é isso?
Se você fizer um gráfico da variável x com a y, os pontos devem formar mais ou
menos uma linha reta (no caso simples), ou um padrão que possa ser ajustado por
um plano (no caso múltiplo).

❌ Se a relação for curva (ex: parabólica, exponencial), a regressão linear vai
errar muito.

👉 Dica: Use gráficos de dispersão ou matriz de correlação para verificar se a
relação é linear.


##2. Ausência de multicolinearidade (no caso da múltipla)##
Multicolinearidade acontece quando duas ou mais variáveis independentes são
muito parecidas entre si.

❌ Problema:
O modelo "se confunde" e não sabe qual variável está de fato influenciando o
resultado.

🛠 Como detectar:
Use a correlação entre variáveis ou métricas como o VIF
(Variance Inflation Factor).


##3. Homoscedasticidade (variância constante dos erros)##
Os erros (diferença entre valor previsto e real) devem ter distribuição parecida
ao longo de todos os valores de x.

❌ Se os erros aumentam ou diminuem com x, o modelo está com problema
(heteroscedasticidade).
🛠 Dica:
Faça um gráfico de resíduos (erros) vs. valores previstos. Eles devem parecer
aleatórios, sem padrão.


##4. Erros devem ser normalmente distribuídos##
Isso é importante principalmente quando usamos regressão para inferência
estatística (como p-valores ou intervalos de confiança).

🛠 Dica:
Faça um histograma ou gráfico Q-Q dos resíduos (erros). Eles devem parecer uma
distribuição normal (curva de sino).

##5. Ausência de outliers extremos
Outliers são valores muito diferentes dos outros, e podem puxar a linha da
regressão para o lado errado.

🛠 Dica:
Use gráficos de dispersão ou boxplots para detectar outliers. Se forem erros ou
exceções, você pode:
* Remover
* Corrigir
* Transformar os dados

##6. Variáveis na mesma escala
Se você tiver variáveis com escalas muito diferentes
(ex: tamanho em m² e preço em milhões), isso pode atrapalhar o aprendizado do
modelo.

🛠 Dica:
Use padronização ou normalização (com StandardScaler, por exemplo).

##7. Quantidade de dados suficiente
Se você tiver poucos dados, o modelo pode se ajustar demais (overfitting) ou
não aprender nada.

🛠 Dica:
Tente ter uma base de dados com amostras representativas e diversas.

##8. Separar dados de treino e teste
Nunca treine e teste o modelo com os mesmos dados.

✅ Correto:
Separe os dados em conjunto de treino e teste
(por exemplo, 80% para treino, 20% para teste).

Assim você testa o desempenho real do modelo em dados novos.


##9. Avaliar o modelo com métricas adequadas
Use métricas específicas para regressão para avaliar a qualidade da previsão.

📊 Métricas comuns:
MSE (Erro Quadrático Médio)
RMSE (Raiz do Erro Quadrático Médio)
MAE (Erro Absoluto Médio)
R² (Coeficiente de Determinação)

##10. Não usar variáveis irrelevantes
Variáveis que não têm relação com y só atrapalham o modelo.

🛠 Dica:
Use análise de correlação, seleção de atributos (feature selection) ou
algoritmos de importância de variáveis para identificar o que deve ou não
entrar no modelo.

````python
# Verificando multicolinearidade (correlação + VIF)
# 2.1 Matriz de correlação
'''
📌 Explicação:
Este gráfico mostra a correlação entre as variáveis.
Correlação alta entre variáveis independentes (ex: tamanho e quartos) pode
indicar problemas de multicolinearidade.
'''
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correlação")
plt.show()


# 2.2 Calculando VIF (Variance Inflation Factor)
'''
📌 Explicação:

O VIF (Variance Inflation Factor) mede quanto uma variável está relacionada com
as outras.
Valores de VIF:
Abaixo de 5: tranquilo
Acima de 5: possível alerta
Acima de 10: grave
'''
X_vif = df[['tamanho_m2', 'quartos', 'distancia_centro']]
X_vif = sm.add_constant(X_vif)

vif = pd.DataFrame()
vif["Variável"] = X_vif.columns
vif["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print("VIF (quanto maior, mais colinearidade):\n", vif)
````

````python
# Verificando homoscedasticidade dos erros
# 3.1 Separando X e y
'''
Explicação:
X = entradas (variáveis independentes), y = saída (preço).
'''
X = df[['tamanho_m2', 'quartos', 'distancia_centro']]
y = df['preco']

# 3.2 Padronizando variáveis
'''
StandardScaler padroniza as variáveis para que fiquem na mesma escala.
'''
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3.3 Separando em treino e teste
'''
Explicação:
Separa os dados:
80% para treinar o modelo
20% para testar depois
'''
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 3.4 Treinando o modelo
'''
Explicação:
Cria e treina o modelo de regressão.
'''
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# 3.5 Previsão
'''
y_pred são os valores previstos.
residuos é o erro: diferença entre valor real e o previsto.
'''
y_pred = modelo.predict(X_test)
residuos = y_test - y_pred

# 3.6 Plotando resíduos vs valores previstos
'''
Explicação:
Este gráfico mostra se os erros são distribuídos de forma uniforme.
Se formarem um "padrão", algo está errado. Se parecerem espalhados
aleatoriamente, está tudo bem!
'''
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_pred, y=residuos)
plt.axhline(0, color='red', linestyle='--')
plt.title("Resíduos vs Valores Previstos (deve parecer aleatório)")
plt.xlabel("Valores previstos")
plt.ylabel("Resíduos")
plt.show()
````

````python
# # Verificar se os erros são normais (forma de sino)
# Histograma
'''
Explicação:
O histograma e o gráfico Q-Q servem para verificar se os erros seguem uma
distribuição normal (curva de sino).
Isso é importante em modelos estatísticos!
'''
plt.hist(residuos, bins=20, edgecolor='k')
plt.title("Distribuição dos Resíduos")
plt.xlabel("Erro")
plt.ylabel("Frequência")
plt.show()

# Q-Q Plot
'''
Explicação:

O histograma e o gráfico Q-Q servem para verificar se os erros seguem uma
distribuição normal (curva de sino).
Isso é importante em modelos estatísticos!
'''
sm.qqplot(residuos, line='s')
plt.title("Q-Q Plot dos Resíduos")
plt.show()

# 4.3 Teste de Shapiro-Wilk
'''
Explicação:
O teste de Shapiro-Wilk testa a normalidade dos erros.
Se o p-valor for maior que 0.05, os resíduos são normalmente distribuídos. ✅
'''
shapiro = stats.shapiro(residuos)
print(f"Shapiro-Wilk p-valor: {shapiro.pvalue:.4f} (ideal > 0.05 para normalidade)")
````

````python
#  Verificando outliers com boxplot
# Boxplots
'''
Explicação:
Os boxplots mostram valores fora do padrão (outliers).
Um ou outro outlier é normal, mas muitos podem prejudicar o modelo.
'''
for coluna in X.columns:
    sns.boxplot(data=df, x=coluna)
    plt.title(f"Boxplot de {coluna} (verificar outliers)")
    plt.show()
````

````python
# Avaliar o modelo com métricas adequadas
'''
Explicação:
MAE: Erro médio absoluto Quanto menor melhor
MSE: Erro médio quadrado (penaliza mais os erros grandes) Quanto menor melhor
RMSE: Raiz do erro quadrado (em R$) Quanto menor melhor
R²: Mostra o quanto o modelo explica dos dados (1 = perfeito)
Escala do R²:
1.00 = modelo perfeito (acertou tudo)
0.00 = modelo inútil (não explica nada)
< 0.00 = pior que a média dos dados (modelo ruim mesmo!)
'''
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
media_preco = df['preco'].mean()
print(f"Média dos preços das casas: R$ {media_preco:.2f}")
print("\nMétricas do modelo:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.4f}")
````
